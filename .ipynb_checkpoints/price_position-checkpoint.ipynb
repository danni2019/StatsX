```python
import utils.database.unified_db_control as udc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = udc.UnifiedControl(db_type='base')
```

```python
sym = 'CF'
exchange='CZCE'
```

```python
md = base.read_dataframe('processed_future_cn_md_data', f'all_1d_main_{exchange}', filter_keyword={'symbol': {'eq': sym}, 'process_type': {'eq': 'O_M_N'}})
```

```python
hv = base.read_dataframe('processed_future_cn_trade_data', f'historical_volatility_{exchange}', filter_keyword={'symbol': {'eq': sym}})
```

```python
hv.sort_values(by='trading_date')
```

```python
md.set_index(['trading_date', 'contract'], inplace=True, drop=False)
hv.set_index(['trading_date', 'contract'], inplace=True)
```

```python
md['hv20'] = hv.loc[hv.index.intersection(md.index)]['hv20']
```

```python
md['rv20'] = hv.loc[hv.index.intersection(md.index)]['rv20']
```

```python
iv_range_df = base.read_dataframe('pretreated_option_cn_md_data', 'all_1d_iv_range_DIY', filter_keyword={'underlying_contract': {'in': md['contract'].drop_duplicates().tolist()}})
```

```python
iv_range_df.set_index(['trading_date', 'underlying_contract'], inplace=True)
```

```python
md['iv_call_avg'] = iv_range_df.loc[iv_range_df.index.intersection(md.index)]['call_avg']
md['iv_call_upper'] = iv_range_df.loc[iv_range_df.index.intersection(md.index)]['call_upper']
md['iv_call_lower'] = iv_range_df.loc[iv_range_df.index.intersection(md.index)]['call_lower']
md['iv_put_avg'] = iv_range_df.loc[iv_range_df.index.intersection(md.index)]['put_avg']
md['iv_put_upper'] = iv_range_df.loc[iv_range_df.index.intersection(md.index)]['put_upper']
md['iv_put_lower'] = iv_range_df.loc[iv_range_df.index.intersection(md.index)]['put_lower']
```

```python
md = md.reset_index(drop=True).set_index('trading_date')
```

```python
md.sort_index(ascending=True, inplace=True)
```

```python
rhigh = md['close'].rolling(window=20).max()
rlow = md['close'].rolling(window=20).min()
md['amp'] = rhigh / rlow - 1
```

```python
md[['close', 'hv20', 'rv20', 'amp']].plot(secondary_y='close', figsize=(16, 9))
```
