```python
import utils.database.unified_db_control as udc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = udc.UnifiedControl(db_type='base')
```

```python
sym = 'AO'
exchange='SHFE'
```

```python
md = base.read_dataframe('processed_future_cn_md_data', f'all_1d_main_{exchange}', filter_keyword={'symbol': {'eq': sym}, 'process_type': {'eq': 'O_M_N'}}, filter_datetime={'trading_date': {'gte': '2024-08-01'}})
```

```python
md = md.set_index('trading_date')[['settlement', 'volume']]
```

```python
bin_width = 50
min_val = md['settlement'].min()
max_val = md['settlement'].max()
bins = range(int(min_val), int(max_val)+bin_width, bin_width)

md['settlement_bin'] = pd.cut(md['settlement'], bins=bins, right=False)
df = md.groupby('settlement_bin')[['volume']].sum()
# df.sort_index(ascending=True, inplace=True)
```

```python
df.plot(kind='barh')
```
